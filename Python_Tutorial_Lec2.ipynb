## **Python Lists**
List is a data structure in which we can store positionally ordered collections of arbitrary object types like integer, float, string etc.

It has a mutable nature.

It has no fixed size.

Duplicates are allowed.

Syntax: list_name = [item 1, item 2, …, item n]
Items or members of list are written inside [ ] separated by comma.



```python
list = ["apple", "banana", "cherry", "apple", "cherry",10,24,2023]
print(list)
print(type(list))
```

## **Operations on List**
**len()**

To count number of elements in a list

**Syntax:** len(list_name)

**append()**

To add items in the list

**Syntax:** list_name.append(item-name or index)

**delete()**

To delete list or specific items of list

**Syntax:** del list_name
del list_name[0]

**index()**

To find index of specific item

**Syntax:** list_name.index(‘item’)

**pop()**

Remove item from specific index

**Syntax:** list_name.pop()

**sort()**

To arrange items of list

**Syntax:**
list_name.sort()

**reverse()**
To reverse the order

**Syntax:**
list_name.reverse()

**count()**

How many time specific item is present is list

**Syntax:** list_name.count(‘item’)

**Slicing**

Slicing specific region inside list

**Syntax:** list_name[:-n]

**Concatenate**

Concatenate two or more lists

Syntax: list1 + list2
list1 + [items..]















## Python Sets
Python’s built-in set type has the following characteristics:
•	 unordered collection of items
•	 Set elements are unique. Duplicate elements are not allowed.
•	A set can have any number of items and there may be different types(integer, float, tuple, string etc.)
•	Since sets are unordered in nature, we cannot apply indexing to them.



```python
set1 = {1,2,2,3,4,5,6,7,"a","g"}
set2 = set([1,2,33,"apple","grapes"])
print(set1)
print(set2)
print(type(set1))
print(type(set2))
```

## **Python Dictionary**
Python dictionary can be defined as a collection of objects, values or items of different types stored in the key-value pair format.
These multiple key-value pairs created are enclosed within the curly braces { },
and each key is separated from its value by the colon (:)

**Syntax:** Dict1 = {1:"a",2:"b",3:"c"}

## **Characteristics of Dictionary**
Elements of the dictionaries can not be accessed by the index.
A dictionary can store various types of elements
Dictionaries are mutable.
Dictionaries don’t allow duplicate elements.



```python
#dictionary
Dict1 ={1:"apple", 2:"banana",3:"grapes"}
print("Dictionary1",Dict1)
print(type(Dict1))

# dictionary using the dict command
Dict2 = dict(name = "John", age = 36, country = "Norway")
print("Dictionary 2",Dict2)
print(type(Dict2))
```

## **Tuple in Python**
Tuple is an immutable collection of elements.
Tuple items are ordered, unchangeable, and allow duplicate values.
In Python, we create tuples by placing the elements inside round brackets and elements separated by commas


```python
# Tuple
tuple = ("apple", "banana",1,2,3, "apple", "cherry")
print(tuple)
print(type(tuple))
```
